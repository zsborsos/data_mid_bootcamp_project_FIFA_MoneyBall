## Data cleaning

After familiarizing myself with the dataframe through excel, and coming up with the questions I would like to explore, I could define which columns I want to drop. After dropping the unwanted columns the dataframe became more manageble.

```python
cleaned = cleaned.drop(['Unnamed: 0', 'Player Photo', 'Club Logo', 'Flag Photo', 'Team & Contract', 'Growth', 'Joined', 
                        'Loan Date End', 'Release Clause', 'Crossing', 'Finishing', 'Heading Accuracy', 'Short Passing',
                        'Volleys', 'Dribbling', 'Curve', 'FK Accuracy', 'Long Passing', 'Ball Control', 'Acceleration',
                        'Sprint Speed', 'Agility', 'Reactions', 'Balance', 'Shot Power', 'Jumping', 'Stamina', 'Strength', 
                        'Long Shots', 'Aggression', 'Interceptions', 'Positioning', 'Vision', 'Penalties', 'Composure', 
                        'Marking', 'Standing Tackle', 'Sliding Tackle', 'W/F', 'SM', 'A/W', 'D/W', 'Hits', 'LS', 'ST', 
                        'RS', 'LW', 'LF', 'LF', 'CF', 'RF', 'RW', 'LAM', 'CAM', 'RAM', 'LM', 'LCM', 'CM', 'RCM', 'RM', 
                        'LWB', 'LDM', 'CDM', 'RDM', 'RWB', 'LB', 'LCB', 'CB', 'RCB', 'RB', 'GK', 'Gender', 'Contract', 
                        'GK Diving', 'GK Handling', 'GK Kicking', 'GK Positioning', 'GK Reflexes', 'Position'], axis=1)
```

```python
cleaned
```

```python
#Standardize header names
cleaned.columns = [column.lower().replace(' ', '_') for column in cleaned.columns]
cleaned
```

```python
#Remove euro sign, remove K
currency_correct = ['value', 'wage']
cleaned[currency_correct] = cleaned[currency_correct].replace({'€':''}, regex=True)
cleaned[currency_correct] = cleaned[currency_correct].replace({'K':'000'}, regex=True)
```

```python
# Deal with M
def money_correct(x):
    if('M' in str(x)):
        x = str(x).replace('M', '')
        x = float(x) * 1000000
    elif('K' in str(x)):
        x = str(x).replace('K', '')
        x = float(x) * 1000
    return float(x)
```

```python
cleaned['value'] = cleaned['value'].apply(money_correct)
```

```python
# Convert value and wage column data types to integers
cleaned = cleaned.astype({"value": 'int64'})
```

```python
cleaned = cleaned.astype({"wage": 'int64'})
```

```python
#Remove lbs from weight
cleaned.weight.unique()
```

```python
weight_correct = ['weight']
cleaned[weight_correct] = cleaned[weight_correct].replace({'lbs':''}, regex=True)
```

```python
cleaned = cleaned.astype({"weight": 'int64'})
```

```python
# Convert height to inches
cleaned.height.unique()
```

```python
height_correct = ['height']
cleaned[height_correct] = cleaned[height_correct].replace({'"':''}, regex=True)
cleaned.height.unique()
```

```python
cleaned.height = [s.split("'") for s in cleaned.height]
cleaned.height = [float(value[0])*12 + float(value[1]) for value in cleaned.height]
```

```python
height_correct = ['height']
cleaned[height_correct] = cleaned[height_correct].replace({'.':''}, regex=True)
cleaned.height.unique()
```

## Outcome 3: Comparison of current German and Hungarian players

```python
query_hun_ger = mysql("SELECT * FROM current_players_with_value WHERE nationality IN ('Hungary', 'Germany')")
query_hun_ger
```

### Visualization of the number of current German and Hungarian players

```python
query_hun_ger_gk = mysql("SELECT * FROM current_players_with_value WHERE nationality IN ('Hungary', 'Germany') AND bp IN ('GK')")
query_hun_ger_gk
```

```python
query_hun_ger_gk['nationality'].value_counts().plot(kind='pie', figsize=(20, 10))
```

### Comparison of average wage of current German and Hungarian players

```python
query_hun_ger_gk_wage = mysql("SELECT nationality, COUNT(nationality) AS 'Number of players', ROUND(AVG(wage)) AS 'Average GK wage' FROM current_players_with_value WHERE nationality IN ('Hungary', 'Germany') GROUP BY nationality")
query_hun_ger_gk_wage
```

### Comparison of average wage of current German and Hungarian goalkeepers

```python
query_hun_ger_wage = mysql("SELECT nationality, COUNT(nationality) AS 'Number of goalkeepers', ROUND(AVG(wage)) AS 'Average wage' FROM current_players_with_value WHERE nationality IN ('Hungary', 'Germany') AND bp IN ('GK') GROUP BY nationality")
query_hun_ger_wage
```
